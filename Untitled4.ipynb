```python
import zipfile
import os

# Specify the name of the uploaded zip file
zip_filename = '/content/drive/My Drive/ecommerce products.zip'

# Specify the directory to extract the contents
extraction_path = '/content/drive/My Drive/'  # Change as needed

# Create the extraction directory if it doesn't exist
os.makedirs(extraction_path, exist_ok=True)

# Extract the contents of the zip file
with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Specify the ecommerce product main directory
main_directory = 'ecommerce products'  # Change this to your actual main directory name

# List of subdirectories within the ecommerce product main directory
subdirectories = ['jeans', 'sofa', 'tshirt', 'tv']  # Replace with your actual subdirectory names

# Loop through each subdirectory
for target_subdirectory in subdirectories:
    # Create a list of image files from the current subdirectory
    image_files = []
    current_subdirectory = os.path.join(extraction_path, main_directory, target_subdirectory)
    for root, dirs, files in os.walk(current_subdirectory):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
                image_files.append(os.path.join(root, file))

    # Display the first image from the current subdirectory as an example
    if image_files:
        print(f"Displaying images from '{target_subdirectory}' subdirectory:")
        first_image_path = image_files[0]
        img = Image.open(first_image_path)
        plt.imshow(img)
        plt.show()
    else:
        print(f"No image files found in '{target_subdirectory}' subdirectory.")
```
